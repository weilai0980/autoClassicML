```python
%matplotlib inline    
import matplotlib as mplt
from matplotlib import cm
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

import matplotlib
from matplotlib.ticker import FuncFormatter
import matplotlib.pyplot as plt
# matplotlib.style.use('ggplot')

from utils_libs import *
from utils_data_prep import *

from scipy.stats import lognorm
from scipy.stats import norm
# from scipy.stats import chisqprob

from numpy import prod
import seaborn as sns

# statiscal models
import statsmodels as sm
from statsmodels.tsa.stattools import acf  
from statsmodels.tsa.stattools import pacf
from statsmodels.tsa.api import VAR, DynamicVAR

from statsmodels.stats import diagnostic
import datetime
import pickle
from datetime import timedelta
import math
import datetime as dt
```

```python
#load data division - tian 

path_data = "../../dataset/bitcoin/double_trx/"

import pickle
tr_dta = pickle.load(open(path_data + 'train.p', "rb"))
val_dta = pickle.load(open(path_data + 'val.p', "rb"))
ts_dta = pickle.load(open(path_data + 'test.p', "rb"))

print(type(tr_dta))
print("loaded")
```

```python
# Training .... data

vol = [ x[0] for x in tr_dta ]

eps = 10**(-15)
#r = [log((j+eps)/(i+eps)) for i, j in zip(vol[:-1], vol[1:])] 
vol_time = [ dt.datetime.utcfromtimestamp(x[1]) for x in tr_dta ]
#vol_series = pd.Series(r, vol_time[1:] )
vol_series = pd.Series(vol, vol_time )
vol_series.plot()
```

```python
# AR + GARCH -- Building a Model From Components

from arch.univariate import ARX
from arch.univariate import ARCH, GARCH

ar = ARX(vol_series, lags = [1, 2, 3, 4, 5])
ar.volatility = GARCH(p=1, q = 1)

# Optional - change distribution
# from arch.univariate import StudentsT
# ar.distribution = StudentsT()

ar_garch_res = ar.fit()
print(ar_garch_res.summary())
fig = ar_garch_res.plot()
```

```python
#evaluation on train data

idx_start = 5
start_date = vol_time[idx_start]
forecasts = ar_garch_res.forecast(horizon=1, start=start_date)

from sklearn.metrics import mean_squared_error
from math import sqrt

mean_pred = np.array(forecasts.mean[idx_start:]).flatten()
true_vol = vol_series[idx_start:].values
error_pred = np.array(np.sqrt(forecasts.variance.values[idx_start:])).flatten()


rms = sqrt(mean_squared_error(mean_pred, true_vol))
observations = true_vol
neg_log_lik = np.sum(log(error_pred**2)*0.5 + log(2*pi)*0.5 +  (1/(2*error_pred**2))*((observations-mean_pred)**2))

rms_check = sqrt(np.sum((true_vol-mean_pred)**2)/len(true_vol))
print(rms_check)


print("RMSE")
print(rms)

print("Neg log likelhood")
print(neg_log_lik)

print("Neg log likelhood normalized")
print(neg_log_lik/len(observations))
```

```python
# test dataset evaluation 
# we construct new ARCH object with fixed parametres from training

vol = [ x[0] for x in ts_dta ]

eps = 10**(-15)
#r = [log((j+eps)/(i+eps)) for i, j in zip(vol[:-1], vol[1:])] 
vol_time = [ dt.datetime.utcfromtimestamp(x[1]) for x in ts_dta ]
#vol_series = pd.Series(r, vol_time[1:] )
vol_series_test = pd.Series(vol, vol_time )

vol_series_test.plot()

print(ar_garch_res.params)

ar_test = ARX(vol_series_test, lags = [1, 2, 3, 4, 5])
ar_test.volatility = GARCH(p=1, q = 1)

#parameters are fixed 
ar_test_res = ar_test.fix(ar_garch_res.params)

print(ar_test_res.summary())

# applying model on the test data.
# no fitting on test data !!!!
# just ally by using forecast method
idx_start = 5
start_date = vol_time[idx_start]
print(start_date)
forecasts_test = ar_test_res.forecast(horizon=1, start=start_date)

forecasts_test.mean.plot()
forecasts_test.variance.plot()
```

```python
# evaluation RMSE neg log likelihood normalzied

from sklearn.metrics import mean_squared_error
from math import sqrt

mean_pred = np.array(forecasts_test.mean[idx_start:]).flatten()
true_vol = vol_series_test[idx_start:].values
error_pred = np.array(np.sqrt(forecasts_test.variance.values[idx_start:])).flatten()

rms = sqrt(mean_squared_error(mean_pred, true_vol))
observations = true_vol
neg_log_lik = np.sum(log(error_pred**2)*0.5 + log(2*pi)*0.5 +  (1/(2*error_pred**2))*((observations-mean_pred)**2))

print("RMSE")
print(rms)

print("Neg log likelhood")
print(neg_log_lik)

print("Neg log likelhood normalized")
print(neg_log_lik/len(observations))

print(ar_test_res.summary())
```

```python
### ADDITIONAL CHECK WHETHER THE RESULTS IS CHANGING IF THE FORECASTING IS DONE BY SPECIFYING THE LAST OBSERVATION DATE


# concatenate train and test -> verification 

vol_train = [ x[0] for x in tr_dta ]
vol_test = [ x[0] for x in ts_dta ]

vol_time_train = [ dt.datetime.utcfromtimestamp(x[1]) for x in tr_dta ]
vol_time_test = [ dt.datetime.utcfromtimestamp(x[1]) for x in ts_dta ]

vol_series_test = pd.Series(vol_train+vol_test, vol_time_train+vol_time_test )

split_date = vol_time_train[-1]
print(split_date)

print(vol_time_test[0])

ar_all = ARX(vol_series_test, lags = [1, 2, 3, 4, 5])
ar_all.volatility = GARCH(p=1, q = 1)

ar_all_res = ar_all.fit(last_obs = split_date)
print(ar_all_res.summary())

forecasts_test = ar_all_res.forecast(start=split_date)

mean_pred = np.array(forecasts_test.mean[split_date:].values).flatten()
true_vol = np.array(vol_series_test[split_date:].values).flatten()
error_pred = np.array(np.sqrt(forecasts_test.variance[split_date:].values)).flatten()

print(mean_pred.shape)
print(true_vol.shape)

forecasts_test.mean.plot()

rms = sqrt(mean_squared_error(mean_pred, true_vol))
observations = true_vol
neg_log_lik = np.sum(log(error_pred**2)*0.5 + log(2*pi)*0.5 +  (1/(2*error_pred**2))*((observations-mean_pred)**2))

print("RMSE")
print(rms)

print("Neg log likelhood")
print(neg_log_lik)

print("Neg log likelhood normalized")
print(neg_log_lik/len(observations))

```
